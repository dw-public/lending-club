```python
import pandas as pd
import json

from IPython.core.display import display, HTML
from sklearn.model_selection import train_test_split, cross_val_score, KFold, StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

from lending_club import db, facets_wrapper, clean

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
data_dict_path = 'data/LCDataDictionary_dw.txt'
csv_path = 'data/lending-club-loan-data/loan.csv'
full_data_dict_path = 'data/LCDataDictionary_dw_types.tsv'
final_data_dict_path = 'data/LCDataDictionary_dw_types_final.txt'

test = False
test_load = 50000

k_fold = 20
```

```python
data_dict = pd.read_csv(final_data_dict_path, sep='\t')
```

```python
# Load the final data dictionary
date_cols = list(data_dict.loc[data_dict.date_format.notnull(),'name'])

# Load the data - only 36 month terms
query = """
SELECT * FROM loan where term=' 36 months';
"""
loan_36_full = db.sql_to_df(query)
del loan_36_full['index']

# Convert issue date to datetime and filter out only loans with full term, after 2012-12-01
loan_36_full.issue_d = pd.to_datetime(loan_36_full.issue_d, format='%b-%Y')
criteria = loan_36_full[(loan_36_full.issue_d <= '2012-12-01')].index
loan_36 = loan_36_full.loc[criteria,:]


#Replace values in categoricals with cleaned, stored in data/json_maps_out.json
with open('data/json_maps_out.json', 'r') as file:
    field_maps = json.load(file)
for column, field_map in field_maps.items():
    loan_36[column] = loan_36[column].map(field_map)
    
# Format the rest of the dates
for col in ['earliest_cr_line', 'last_credit_pull_d']:
    loan_36[col] = pd.to_datetime(loan_36[col], format='%b-%Y')
    
# Convert int_rate to numeric
loan_36.int_rate = clean.percent_to_float(loan_36.int_rate)

# Convert revol_util to numeric
loan_36.revol_util = clean.percent_to_float(loan_36.revol_util)

loan_36.set_index('id', inplace=True)
print(loan_36_full.shape)
print(loan_36.shape)
loan_36.head()
```

# Cleaning and Feature Selection

```python
potential_feature_list = clean.columns_by_type(data_dict, 'demographics') + clean.columns_by_type(data_dict, 'loan') + clean.columns_by_type(data_dict, 'credit') + clean.columns_by_type(data_dict, 'LC_calcs')

# Create features, using only variables available at origination
potential_feature_list = clean.columns_by_type(data_dict, 'demographics') + clean.columns_by_type(data_dict, 'loan') + clean.columns_by_type(data_dict, 'credit') + clean.columns_by_type(data_dict, 'LC_calcs')

# Eliminate url, desc, emp_title, and title since unstructured text fields. Could parse into stemmed terms in future iteration
for x in ['url', 'desc', 'emp_title', 'title', 'id', 'member_id', 'addr_state']:
    potential_feature_list.remove(x)
```

```python
html = facets_wrapper.make_facet(loan_36[potential_feature_list], 'name')
display(HTML(html))
```

#### Notes
##### Numeric
* Most of the detailed credit information is missing for these loans. Lending Club must have only started collecting it post 12/2012.
* None of the loans are joint. Perhaps not possible in 36 month terms or not offered until post 2012.
* Employment length should probably be binned and then encoded
* Annual income should probably be binned and then encoded
* delinq_2yrs, inq_last_6mths have a high percentage of 0s (87% and 49% respectively). Add a binary classification feature. Maybe a bin?
* mnths_since_last_delinq, mnths_since_last_record, and mths_since_last_major_derog have a high percentage of missing (60.59%, 94.41%, 94.09% respectively). The numbers here seem important, since mths_since_last_record also has a high percentage of 0s (30.39%). The people with zeros have a record in the same month, which sounds bad. Let's bin and encode.
* Revol_bal should be binned, similar to salary
* tot_coll_amt, tot_cur_bal, and tot_rev_hi_lim all have high percentage of missing data (72.11%) Binarize or bin and encode.
* installment, int_rate, and loan_amnt are correlated. Should probably only use one.
* loan amount has a moderate tail to the right - consider log(loan amount)

##### Text
* drop application_type and term
* zipcode and state are reflective of broader economic trends. Zip in itself is probably too detailed to use without mapping to something like median salary. State may be useable as is (w/ encoding)

```python
# Create potential features table, loan_36_pf
loan_36_pf = loan_36[potential_feature_list+['loan_status']].copy()
print(loan_36_pf.shape)

# Drop columns where all values are NA - will get rid of all the missing credit vars
print('Drop columns where all values are NA')
loan_36_pf.dropna(axis=1, how='all', inplace=True)
print(loan_36_pf.shape)

# Drop columns where all rows have the same value
for col in loan_36_pf.columns:
    if loan_36_pf[col].value_counts().shape[0]==1:
        print('Dropping column {column} because single value.'.format(column=col))
        del loan_36_pf[col]
print(loan_36_pf.shape)

del loan_36_pf['zip_code']

# NEW FEATURES
#-------------

## LEN CREDIT HISTORY
loan_36_pf['len_credit_history'] = loan_36_pf.apply(lambda x: clean.difference_in_months(x.last_credit_pull_d, x.earliest_cr_line), axis=1)

### eliminate earliest_credit_line and last_credit_pull_date
del loan_36_pf['last_credit_pull_d']
del loan_36_pf['earliest_cr_line']

## mnths_since_last_delinq, mnths_since_last_record, and mths_since_last_major_derog
# Use missing indicator method (make variable to indicate missing, replace missing val with another number)
for col in ['mths_since_last_delinq', 'mths_since_last_record', 'mths_since_last_major_derog']:
    loan_36_pf[col+'_missing']=loan_36_pf[col].isnull().astype(int)
    loan_36_pf[col].fillna(999, inplace=True)
    
## tot_coll_amt, tot_cur_bal, and tot_rev_hi_lim, emp_length - replace with 0s
for col in ['tot_coll_amt', 'tot_cur_bal', 'total_rev_hi_lim', 'emp_length']:
    loan_36_pf[col+'_missing']=loan_36_pf[col].isnull().astype(int)
    loan_36_pf[col].fillna(0, inplace=True)

loan_36_pf.shape
```

```python
html = facets_wrapper.make_facet(loan_36_pf, 'name')
display(HTML(html))
```

```python
# Drop rows with missing vals
loan_36_pf.dropna(axis=0, how='any', inplace=True)
loan_36_pf.shape
```

```python
# Handle categoricals

# Drop rows that are late, current, and in default bc don't know what will happen with them.
loan_36_pf = loan_36_pf[~(loan_36_pf.loan_status.isin(['late', 'current', 'default']))]
print(loan_36_pf.shape)

# Convert charged off and fully paid to binary outcome
loan_36_pf.loan_status.replace({'charged_off':1, 'fully_paid':0}, inplace=True)

# Expand categoricals
cat_vars = ['home_ownership', 'purpose', 'grade', 'initial_list_status', 'sub_grade', 'verification_status']
for var in cat_vars:
    cat_list = 'var'+'_'+'var'
    cat_list = pd.get_dummies(loan_36_pf[var], prefix=var)
    tmp_df = loan_36_pf.join(cat_list)
    loan_36_pf = tmp_df

pf_feature_vars = loan_36_pf.columns.values.tolist()
to_keep = [i for i in pf_feature_vars if i not in cat_vars]
# post process
loan_36_pp = loan_36_pf[to_keep]
print(loan_36_pp.shape)
loan_36_pp.head()
```

## Train/Test Split

```python

X = loan_36_pp.drop('loan_status', axis=1).sort_index()
y = loan_36_pp['loan_status'].sort_index()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

# Feature Selection

```python
logreg = LogisticRegression()

rfecv = RFECV(logreg, step=1, cv=StratifiedKFold(k_fold), scoring='roc_auc', n_jobs=-1)
rfecv.fit(X_train,y_train)

print('Optimal number of features: {n}'.format(n=rfecv.n_features_))

plt.figure()
plt.xlabel('Number of features selected')
plt.ylabel('Cross validation score (nb of correct classifications)')
plt.plot(range(1, len(rfecv.grid_scores_)+1), rfecv.grid_scores_)
plt.show()
```

```python
cv_features = X_train.columns[rfecv.support_]
print(cv_features)
```

```python
X_train_f = X_train[cv_features]
X_test_f = X_test[cv_features]
```

# Model Fit

```python
logreg_f = LogisticRegression()
logreg_f.fit(X_train_f, y_train)
```

Use the accuracy score metric from `sklearn` to assess performance on training and test set. The `accuracy_score` function computes the fraction of correct predictions.

```python
y_train_pred = logreg_f.predict(X_train_f)
print('The accuracy score of the model on the full training data set is {score}.'.format(score=round(accuracy_score(y_train, y_train_pred),3)))
```

```python
y_test_pred = logreg_f.predict(X_test_f)
print('The accuracy score of the model on the reserved test data set is {score}.'.format(score=round(accuracy_score(y_test, y_test_pred),3)))
```
